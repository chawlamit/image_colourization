```python
%load_ext autoreload
%autoreload 2
from ann.activations import LeakyRelu, Sigmoid, Tanh
from ann.linear import Linear

# from ann.sampling import Sampling
from ann.losses import MSE, Loss, BCE, CCE
import numpy as np
from ann.sequential_model import Sequential
from ann.utils import create_dataset, mean_centring
from sklearn.preprocessing import scale

import time
```

```python
data = np.genfromtxt('winequality-red.csv', delimiter=',')[1:]
# Deleting citric acid feature
data = np.delete(data, 2, 1)
# data = data[:, :5]
data = mean_centring(data)
data = scale(data)
x_train, x_test, x_test_with_missing, mask = create_dataset(data)

model = Sequential()
# inputs = keras.Input(shape=(11))
model.add(Linear(11, 9))
model.add(Tanh())

model.add(Linear(9, 7))
model.add(Tanh())

model.add(Linear(7, 5))
model.add(Tanh())

model.add(Linear(5, 7))
model.add(Tanh())

model.add(Linear(7, 9))
model.add(Tanh())

model.add(Linear(9, 11))
```

```python
err = MSE()
model.train(x_train,x_train, err, 2000, batch_size=0, lr=0.000_0005)
```

```python
idx = 1
x_hat = model.forward(x_train)
print(np.mean(err.loss(x_hat, x_train)))
print(x_train[idx])
print(x_hat[idx])
print()
x_test_missing_pred = model.forward(x_test_with_missing)
print(x_test_missing_pred[idx])
print(x_test[idx])
```

```python
np.mean(np.where(mask == 0, (x_test_missing_pred - x_test)**2, 0))
```
